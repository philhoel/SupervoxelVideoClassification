```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from moviepy.editor import VideoFileClip

import sys
sys.path.append("/home/philip/Documents/Master/MasterThesisCode/SupervoxelVideoClassification/src")

from supervoxel.sp_algorithm import Supervoxel
```

```python
clip = VideoFileClip("v_ApplyEyeMakeup_g01_c01.avi")
clip = clip.resize(height=240, width=320)
vid = np.array([frame for frame in clip.iter_frames()])

vid = vid / 255



t = torch.tensor(vid, dtype=torch.float32)
t = t.unsqueeze(0)
t = t.permute(0, -1, 1, 2, 3)

sv = Supervoxel()
sv.process(t, 8)
```

```python
sv_indices = sv.segs.view(-1).mul(sv.segs.shape[0]).add(sv.coords[0]).unique() % sv.segs.shape[0]
print(sv_indices)
```
